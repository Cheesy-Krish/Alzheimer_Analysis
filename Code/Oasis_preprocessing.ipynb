```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.io as io
import itk                                                                # Jupyter widgets to visualize images in 2D and 3D
import itkwidgets
from scipy import ndimage
from ipywidgets import interact, interactive, IntSlider, ToggleButtons
import cv2
import pywt
```

```python
data_path="OAS2_RAW_PART1\OAS2_0001_MR1\RAW"
```

```python
example_filename = os.path.join(data_path, 'mpr-1.nifti.img')
```

```python
#img = nib.load(example_filename)
img = nib.load(example_filename)
```

```python
img.shape
```

```python
image_data = img.get_fdata()
```

```python
height, width, depth ,channel= image_data.shape
print(f"The image object has the following dimensions: height: {height}, width:{width}, depth:{depth},channel:{channel}")
```

```python
#ndimage.rotate(img, 270)
#plt.imshow(image_data[:, :, 100,0], cmap='gray')
plt.subplot(1,2,1)
image=ndimage.rotate(image_data[:, :, 64,0],90)
ret, thresh3 = cv2.threshold(image, 120, 255, cv2.THRESH_TOZERO)
c=pywt.wavedec2(thresh3[70:150,70:170],'db5',mode='periodization',level=2)
#image = cv2.cvtColor(thresh3, cv2.COLOR_RGB2BGR)
cA2=c[0]
(cH1,cV1,cD1)=c[-1]
(cH2,cV2,cD2)=c[-2]
plt.imshow(cH2, cmap='gray')
plt.axis('off');


plt.subplot(1,2,2)
plt.imshow(ndimage.rotate(image_data[:, :, 64,0],90), cmap='gray')
plt.axis('off');
```
